Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?

Answer:- Dimensionality reduction means reducing the number of input features while keeping the important information.

**Why it’s important:**

1. Makes models faster.
2. Reduces overfitting.
3. Removes irrelevant or duplicate features.
4. Helps visualize data in 2D or 3D.


Question 2: Name and briefly describe three common dimensionality reduction
techniques.

Answer:- **1. PCA (Principal Component Analysis)**
Reduces features by finding new directions (components) that capture the most variance in the data.

**2. t-SNE (t-Distributed Stochastic Neighbor Embedding)**
Used mainly for visualization; keeps similar points close together in 2D or 3D.

**3. LDA (Linear Discriminant Analysis)**
Reduces dimensions by maximizing class separation; works for labeled data.

Question 4: Explain the concept of anomaly detection and its significance.

Answer:- Anomaly detection means identifying data points that are unusual or different from the normal pattern.

**Significance:**

1. Helps detect fraud (banking, transactions).
2. Finds faults or failures in machines.
3. Identifies unusual behavior in networks or systems.
4. Improves security and reliability.


Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:- **1. Statistical Methods**
Detect anomalies by checking if a data point is far from the normal distribution (mean, variance, z-score).

**2. Machine Learning Methods**
Use models like Isolation Forest or One-Class SVM to learn normal patterns and flag anything unusual.

**3. Distance/Clustering Methods**
Points far from clusters or with large distances (e.g., using k-NN or DBSCAN) are marked as anomalies.


Question 6: What is time series analysis? Mention two key components of time series data.

Answer:- Time series analysis is the study of data collected over time to understand patterns and make predictions.

Two key components:

1. Trend – long-term upward or downward movement.
2. Seasonality – repeating patterns at regular intervals.



Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:- **Seasonality:**
Repeating patterns that occur at fixed, regular intervals (daily, weekly, yearly).

**Cyclic behavior:**
Patterns that rise and fall but with **no fixed period**; cycles depend on external factors like economy or market conditions.



Question 8: Write Python code to perform K-means clustering on a sample dataset.


```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create sample data
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Plot results
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.


```python
class Animal:
    def speak(self):
        return "Animal sound"

class Dog(Animal):
    def speak(self):
        return "Bark"

d = Dog()
print(d.speak())  # Output: Bark
```

Question 10: How can time series analysis be used for anomaly detection?

Answer:- Time series analysis can detect anomalies by checking if a data point deviates from the normal pattern over time.

Common ways:

1. Compare data to expected trend or seasonality.
2. Use forecasting models (like ARIMA or LSTM) and flag points where actual values differ a lot from predictions.
3. Detect sudden spikes, drops, or unusual patterns.
